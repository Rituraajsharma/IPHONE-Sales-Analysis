# Q.1 What are the top 10 highest-rated iphones on flipkart in india


```python
df.head()
```

```python
highest_rated=df.sort_values(by=["Star Rating"],ascending = False)

highest_rated=highest_rated.head(10)
```

```python
print (highest_rated["Product Name"])
```

```python
print (highest_rated)
```

# Q.2 How many ratings do the highest-rated iphones on flipkart have

```python
df.head()
```

```python
iphones=highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Ratings"]
figure = px.bar(highest_rated,x=label,y=counts,title="Number of reviews of highest rated iphones")
figure.show()
```

#Q.3 Which iphones has the highest number of reviews on flipkart

```python
df.head()
```

```python
iphones=highest_rated["Product Name"].value_counts()
label = iphones.index
counts = highest_rated["Number Of Reviews"]
figure = px.bar(highest_rated,x=label,y=counts,title="Number of reviews of highest rated iphones")
figure.show()
```
